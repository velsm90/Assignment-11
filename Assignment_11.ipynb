```python
import numpy as np
import pandas as pd

url='https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic = pd.read_csv(url)
titanic.head()
```

```python
# 1. Create a pie chart presenting the male/female proportion
import matplotlib as mpl
import matplotlib.pyplot as plt

#titanic['sex'].male.count()
#t = pd.value_counts(titanic['sex'])

# Get total count 
total = titanic['sex'].count()

# Load only male details to new data frame
male = titanic.loc[titanic['sex'] == 'male']
male_count = (male['sex'].count()/total) # get count of males

# Load only female details to new data frame
female = titanic.loc[titanic['sex'] == 'female']
female_count = (female['sex'].count()/total) # get count of females

# Pie chart attributes
labels = ['Male', 'Female']
sizes = [male_count, female_count]
colors = ['Green', 'Orange']
explode = (0, 0.075)

# plotting and displaying of graph
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

## Alternate option to plot charts
#patches, texts = plt.pie(sizes,explode=explode, colors=colors, autopct='%1.1f%%',shadow=True, startangle=90)
#plt.legend(patches, labels, loc="best")
#plt.axis('equal')
#plt.tight_layout()
#plt.show()
```

```python
# 2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

# Load male and female records separatly in new dataframe
male_records = titanic.loc[titanic['sex'] == 'male']
female_records = titanic.loc[titanic['sex'] == 'female']

# plotting scatter plots for male and females
ax1 = male_records.plot(kind='scatter', x='age', y='fare', color='r',label='Male')    
ax2 = female_records.plot(kind='scatter', x='age', y='fare', color='b',label='Female', ax=ax1)
ax1.set_xlabel("Age")
ax1.set_ylabel("Fare")
print(ax1 == ax2)  # Display male and female graph in a single plot

#fig, male_g = plt.subplots()
#myplot = male_g.scatter(male_records["age"],male_records["fare"])
#male_g.set_xlabel("age")
#male_g.set_ylabel("fare")
#male_g.set_title("scatter plot")
#plt.show()
```
